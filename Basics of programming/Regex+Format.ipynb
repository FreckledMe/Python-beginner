```python
#ID   Familiya  Ism   Yoshi  Kursi  Tel   jadval.txt
file_name=r"jadval.txt"

def AddIDnumber():
    global file_name
    with open(file_name,'a') as jadval:
        Fam=input("Familiyani kiriting = ")
        Ism=input("Ismni kiriting = ")
        Yosh=input("Yoshini kiriting = ")
        Kurs=input("Kursini kiriting = ")
        Average=float(input("O'rtacha bahosini kiriting = "))
        Tel=input('Telefoni raqamini kiriting = ')
        print(Fam,"\t",Ism,"\t",Yosh,"\t",Kurs,"\t",Average,"\t",Tel, file=jadval)

def DeleteIDnumber():
    global file_name
    Fam=input("Familiyani kiriting = ")
    with open(file_name,"r") as jadval:
        lines=jadval.readlines()
    with open(file_name,"w") as jadval:
        for i in lines:
            if not i.strip("\n").startswith(Fam):
                jadval.write(i)
    
def EditIDnumber():
    global file_name
    Fam=input("Familiyani kiriting = ")
    with open(file_name,"r") as table:
        lines=table.readlines()
    with open(file_name,"w") as table:
        for i in lines:
            if not i.strip("\n").startswith(Fam):
                table.write(i)
            else:
                print(i)
                Fam=input("Familiyani kiriting = ")
                Ism=input("Ismni kiriting = ")
                Yosh=input("Yoshini kiriting = ")
                Kurs=input("Kursini kiriting = ")
                Average=float(input("O'rtacha bahosini kiriting = "))
                Tel=input('Telefoni raqamini kiriting = ')
                table.write(Fam+"\t"+Ism+"\t"+Yosh+"\t"+Kurs+"\t"+str(Average)+"\t"+Tel+"\n")

def SearchbyFirstName():
    global file_name
    Fam=input("Familiyani kiriting = ")
    with open(file_name,"r") as table:
        for i in table:
            S=i.split('\t')[0]
            if Fam in S:
                print(i)
    
def Searchbyphone():
    global file_name
    Tel=input("Familiyani kiriting = ")
    with open(file_name,"r") as table:
        for i in table:
            S=i.split('\t')[-1]
            if Tel in S:
                print(i)

def Printing():
    global file_name
    print("Familiya\tIsm\tYosh\tKurs\tAverage\tTelefon")
    with open(file_name,"r") as table:
        f=table.read()
        print(f)
    
while True:
    print("Buyruqlar","1-Id qo'shish","2-Idni o'chirish","3-Idni edit qilish","4-Familiya qidirish","5-Telefon qidirish",\
         "6-Ekranga chiqarish","0-Exit",sep='\n')
    buyruq=int(input('Buyruq raqamini tanlang = '))
    if buyruq==1:
        AddIDnumber()
    if buyruq==2:
        DeleteIDnumber()
    if buyruq==3:
        EditIDnumber()
    if buyruq==4:
        SearchbyFirstName()
    if buyruq==5:
        Searchbyphone()
    if buyruq==6:
        Printing()
    if buyruq==0:
        print("Raxmat Men ishlashdan to'xtadim")
        break
```

```python
#Student ID
f_name=r'guruh.txt'
name_f = r'parents.txt'
file_name=r'semestr.txt'
guruhlar_table = []
talabalar_table = []
groups = int(input("Guruhlar sonini kiriting: "))

#Talabalarni kiritib chiqish
def AddIDmember():
    global f_name,groups,guruhlar_table,talabalar_table
    with open(f_name,'a') as member:
        for j in range(groups):
            n = int(input(f"{j+1}-guruh raqamini kiriting:"))
            guruhlar_table.append(n)
            son = int(input("Talabalar sonini kiriting:"))
            talabalar_table.append(son)
            print(f"{n}-guruhdagi talabalar ro'yxati")
            for i in range(son):
                Fam = input(f'{i+1}-talaba familiyasini kiriting:   ')
                Name = input(f'{i+1}-talaba ismini kiriting:   ')
                Kurs=input(f"{i+1}-talaba kursini kiriting:  ")
                Yosh=int(input(f'{i+1}-talaba yoshini kiriting:  '))
                Tel=input(f'{i+1}-talaba telefoni raqamini kiriting:  ')
                print(Fam,'\t',Name,'\t',Kurs,'\t',str(Yosh),'\t',Tel,file=member)
            print('\n',file=member)
            
# Talabalar ro'yxatini qaytaruvchi funksiya
def royxat_qaytar():
    global f_name
    with open(f_name) as guruh:
        royxat = guruh.readlines()
    royxat = [i.replace('\t','',1) for i in royxat]
    royxat = [i[:i.find('\t')] for i in royxat]   
    while royxat.count(''):
        royxat.remove('')
    return royxat     
            
# Talabalarning telefon raqamini qaytaruvchi funksiya
def nomer():
    global f_name
    with open(f_name) as guruh:
        royxat = guruh.readlines()    
    royxat = [i[i.rfind('\t')+1:] for i in royxat]   
    return royxat 

# Talabalarni semestr baholarini qo'yish
def AddIDsemestr():
    global file_name,f_name,guruhlar_table,talabalar_table   
    with open(file_name,'a') as semestr:
        for sem in range(2):
            sanoq = 0
            for i in range(len(guruhlar_table)):
                print(f'{guruhlar_table[i]}-guruh talabalari {sem+1}-semestr baholari')
                for j in range(talabalar_table[i]):
                    print(royxat_qaytar()[sanoq])
                    sanoq += 1
                    baho = int(input('bahosi: '))
                    print(baho,file=semestr)
                    
# Talabalarning baholarini qaytaruvchi funksiya                   
def qator_qaytar():
    global file_name
    with open(file_name) as f:
        qatorlar = f.readlines()
    qatorlar = [int(i[0]) for i in qatorlar]
    return qatorlar

# Talabalarning ota-onalarini qo'shish
def AddIDparents():
    global file_name,f_name,guruhlar_table,talabalar_table   
    sanoq = 0
    with open(name_f,'a') as parents:
        for i in range(len(guruhlar_table)):
            print(f'{guruhlar_table[i]}-guruh')
            for j in range(len(talabalar_table)):
                print(royxat_qaytar()[sanoq])
                sanoq += 1
                ota = input('Otasining ism-familiyasi: ')
                ona = input('Onasining ism-familiyasi: ')
                print(ota,'\t',ona,file=parents)    
                
# Talabalarning ota - onalarini qaytaruvchi funksiya
def parents():
    global name_f
    with open(name_f) as guruh:
        royxat = guruh.readlines()
    royxat = [i.strip('\n').replace('\t',' ') for i in royxat]
    return royxat 
                
#  Guruhlardagi har bir semestr bo'yicha eng kuchli va eng qoloq talabalar       
def strong_semestr():
    global file_name,guruhlar_table,talabalar_table
    hisob = 0
    for semlar in range(2):
        new_royxat1 = royxat_qaytar()
        new_royxat2 = royxat_qaytar()
        qatorlar = qator_qaytar()
        for g in range(len(guruhlar_table)):
            semestr_first = qatorlar[:talabalar_table[g]]
            qatorlar = qatorlar[talabalar_table[g]:]
            hisob += 1

            print(f"{semlar+1}-semestr bo'yicha {guruhlar_table[g]}-guruhdagi eng kuchli talabalar: ")
            for i in range(len(semestr_first)):
                if semestr_first[i] == max(semestr_first):
                    print(f'{new_royxat1[i]} va bahosi: {max(semestr_first)}')
            new_royxat1 = new_royxat1[talabalar_table[g]:]
            
            print(f"{semlar+1}-semestr bo'yicha {guruhlar_table[g]}-guruhdagi eng qoloq talabalar: ")
            for i in range(len(semestr_first)):
                if semestr_first[i] == min(semestr_first):
                    print(f'{new_royxat2[i]} va bahosi: {min(semestr_first)}')
            new_royxat2 = new_royxat2[talabalar_table[g]:]
            
# Umumiy eng kuchli va eng qoloq talabalar,raqamlari va ota - onalarining ism - familiyalari
def umumiy_kuchli():
    global file_name,royxat,guruhlar_table,talabalar_table
    k = 2 * len(guruhlar_table) * len(talabalar_table)
    qadam = sum(talabalar_table)
    sum_massiv = []
    for i in range(len(royxat)):
        sum_massiv.append(sum([qator_qaytar()[i] for i in range(i,len(qator_qaytar()),qadam)])/2)
         
    print('Umumiy eng kuchli talabalar:')    
    for i in range(len(sum_massiv)):
        if sum_massiv[i] == max(sum_massiv):
            print(f'{royxat[i]} va bahosi: {max(sum_massiv)} {nomer()[i]} {parents()[i]}')
            
    print('Umumiy eng qoloq talabalar:')
    for i in range(len(sum_massiv)):
        if sum_massiv[i] == min(sum_massiv):
            print(f'{royxat[i]} va bahosi: {min(sum_massiv)} {nomer()[i]} {parents()[i]}')
            
# Har bir semestrdagi eng kuchli va eng qoloq talabalar ro'yxati           
def har_semestr():
    global talabalar_table,guruhlar_table
    
    total_students = sum(talabalar_table)
    qatorlar_1 = qator_qaytar()
    for j in range(2):
        maxi_sem = qatorlar_1[:total_students]
        qatorlar_1 = qatorlar_1[total_students:]
        
        print(f"{j+1}-semestr bo'yicha eng kuchli talabalar:")
        for i in range(len(maxi_sem)):
            if maxi_sem[i] == max(maxi_sem):
                print(f'{royxat_qaytar()[i]} va bahosi: {max(maxi_sem)}')
                
        print(f"{j+1}-semestr bo'yicha eng qoloq talabalar:")
        for i in range(len(maxi_sem)):
            if maxi_sem[i] == min(maxi_sem):
                print(f'{royxat_qaytar()[i]} va bahosi: {min(maxi_sem)}')
                
AddIDmember()
AddIDsemestr()
AddIDparents()
strong_semestr()
umumiy_kuchli()
```

```python
#avtomobil raqami
import re
pattern_jismoniy = r'^\d{2}[A-Z]\d{3}[A-Z]{2}$'
pattern_yuridik = r'^\d{2}\d{3}[A-Z]{3}$'
raqam = input('Raqam kiriting: ')
def detector(raqam):
    if re.match(pattern_jismoniy,raqam):
        print('Jismoniy shaxs')
    elif re.match(pattern_yuridik,raqam):
        print('Yuridik shaxs')
    else:
        return 'Bunday raqam mavjud emas'
    if raqam[0] == '0':
        print('Toshkent')
    elif raqam[0] == '1':
        print('Toshkent viloyati')
    elif 20<=int(raqam[:2])<=24:
        print("Sirdaryo")
    elif 25<=int(raqam[:2])<=29:
        print("Jizzax")
    elif 30<=int(raqam[:2])<=39:
        print("Samarqand")
    elif 40<=int(raqam[:2])<=49:
        print("Farg'ona")
    elif 50<=int(raqam[:2])<=59:
        print("Namangan")
    elif 60<=int(raqam[:2])<=69:
        print("Andijon")
    elif 70<=int(raqam[:2])<=74:
        print("Qashqadaryo")
    elif 75<=int(raqam[:2])<=79:
        print("Surxondaryo")
    elif 80<=int(raqam[:2])<=84:
        print("Buxoro")
    elif 85<=int(raqam[:2])<=89:
        print("Navoiy")
    elif 90<=int(raqam[:2])<=94:
        print("Xorazm")
    elif 95<=int(raqam[:2])<=99:
        print("Qoraqalpog'iston")
detector(raqam)
```
